# 02 — Revenue Analysis

**Goal:** Analyse total revenue, monthly trends, growth rates, average order value,
and payment method distribution.

---

```python
import sys
sys.path.append('../src')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns

from analysis import (
    set_style,
    monthly_revenue,
    plot_monthly_revenue,
    plot_payment_methods,
    print_kpi_summary,
    PALETTE,
)

set_style()
print('Ready ✅')
```

```python
master = pd.read_csv('../data/processed/master.csv', parse_dates=['order_purchase_timestamp'])
master['order_yearmonth'] = master['order_purchase_timestamp'].dt.to_period('M')
print(f'Loaded {master.shape[0]:,} rows')
```

## 1. Top-Level KPIs

```python
print_kpi_summary(master)
```

## 2. Monthly Revenue Table

```python
monthly = monthly_revenue(master)
monthly.style.format({
    'revenue': 'R${:,.2f}',
    'avg_order_value': 'R${:,.2f}',
    'revenue_growth_pct': '{:+.1f}%'
})
```

## 3. Monthly Revenue Trend

```python
plot_monthly_revenue(monthly, save_path='../images/monthly_revenue.png')
```

## 4. Monthly Orders vs Revenue (Dual Axis)

```python
fig, ax1 = plt.subplots(figsize=(14, 5))

ax1.bar(monthly['order_yearmonth'], monthly['orders'],
        color=PALETTE['secondary'], alpha=0.5, label='Orders')
ax1.set_ylabel('Number of Orders', color=PALETTE['secondary'])
ax1.tick_params(axis='y', labelcolor=PALETTE['secondary'])

ax2 = ax1.twinx()
ax2.plot(monthly['order_yearmonth'], monthly['revenue'],
         color=PALETTE['primary'], marker='o', linewidth=2.5, label='Revenue')
ax2.set_ylabel('Revenue (R$)', color=PALETTE['primary'])
ax2.tick_params(axis='y', labelcolor=PALETTE['primary'])
ax2.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'R${x/1e3:.0f}k'))

ax1.set_title('Monthly Orders vs Revenue', fontsize=15, fontweight='bold')
plt.xticks(rotation=45, ha='right')
fig.tight_layout()
plt.savefig('../images/orders_vs_revenue.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 5. Payment Method Analysis

```python
plot_payment_methods(master, save_path='../images/payment_methods.png')
```

## 6. Average Order Value Over Time

```python
fig, ax = plt.subplots(figsize=(14, 4))
ax.plot(monthly['order_yearmonth'], monthly['avg_order_value'],
        color=PALETTE['accent'], marker='s', linewidth=2)
ax.set_title('Average Order Value (AOV) Over Time', fontsize=14, fontweight='bold')
ax.set_ylabel('R$')
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'R${x:,.0f}'))
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('../images/aov_trend.png', dpi=150, bbox_inches='tight')
plt.show()
```
